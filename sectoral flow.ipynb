```python
# New dict where nifty pharma is deleted
sector_symbols =  ['^NSEBANK', '^CNXAUTO', "NIFTY_FIN_SERVICE.NS",'^CNXFMCG',
                '^CNXIT','^CNXMEDIA','^CNXMETAL','^CNXPSUBANK','^CNXREALTY','^CNXENERGY']

sector_names = ['Nifty Bank','Nifty Auto','Nifty Financial Services','Nifty FMCG',
               'Nifty IT','Nifty Media','Nifty Metal','Nifty PSU Bank',
               'Nifty Realty','Nifty Energy']

# sector index {'symbol' : name}
si = dict(zip(sector_symbols,sector_names))
```

```python
def chngper(dt):
    sidf = pd.DataFrame()
    for i in tqdm(range(len(dt)),desc='Converting Data Into DataFrame'):
        data = yf.download(list(dt.keys())[i],period='2d',progress=False)
        data = data.reset_index()
        perchng = ((data['Adj Close'][1] - data['Adj Close'][0]) / data['Adj Close'][0]) * 100
        sidf.loc[0,list(dt.values())[i]] = perchng
    data = yf.download(list(dt.keys())[i],period='2d',progress=False)
    latest_date = str(data.index[1]).split()[0]
    return sidf,latest_date
```

```python
sidf,latest_date = chngper(si)
```

```python
latest_date
```

```python
# Plotting the horizontal bar chart
ax = sidf.plot.barh(figsize=(12, 6), legend=False,width=2)

# Add labels to the bars
for p in ax.patches:
    width = p.get_width()
    plt.text(width, p.get_y() + p.get_height() / 2, f'{width:.2f}', va='center',fontsize=12)

# Set the axis labels and the title
plt.xlabel('Percentage Change', backgroundcolor='white')
plt.ylabel('Sectors', backgroundcolor='white')
plt.title('Sectoral Flow ( Percentage Change % ) - '+latest_date, backgroundcolor='white',pad=10)

legend = plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=12,facecolor='white')

plt.gca().tick_params(axis='x', colors='blue')
plt.gca().tick_params(axis='y', colors='blue')
plt.figure(facecolor='white')
# Show the plot
plt.show()
```
