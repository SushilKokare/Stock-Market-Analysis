```python
import pandas as pd
from nsepy.history import get_history
from datetime import date
import datetime
```

```python
flag=0
edate=date.today()
df1=pd.DataFrame()
len(df1)
```

```python
while(len(df1)<15):
    flag=flag+1
    sdate=date.today()-datetime.timedelta(flag)
    df1=get_history('sbin',sdate,edate)
```
