```python
import yfinance as yf
from matplotlib import pyplot as plt
```

```python
syear = input('Enter start year (yyyy) ')
smonth = input('Enter start month (mm) ')
sdate = input('Enter start date (dd) ')

start_date =  syear+'-'+smonth+'-'+sdate
start_date
```

```python
eyear = input('Enter end year (yyyy) ')
emonth = input('Enter end month (mm) ')
edate = input('Enter end date (dd) ')

end_date =  eyear+'-'+emonth+'-'+edate
end_date
```

```python
symb = input('Entet stock symbol : ')
symb = symb+'.ns'
```

```python
stock_data = yf.download(symb,start=start_date,end=end_date)
```

```python
stock_data['Daily_Return'] = (stock_data['Adj Close'].pct_change())*100
```

### Daily Return = ( Today’s Adj Close − Yesterday’s Adj Close ) / Yesterday’s Adj Close

```python
len(stock_data)
```

```python

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(stock_data['Daily_Return'], bins=20, edgecolor='black', color='skyblue')
plt.xlabel('Daily Returns')
plt.ylabel('Frequency')
plt.title('Histogram of Daily Returns for {}'.format(symb))
plt.grid(True)

# Show the histogram
plt.show()
```

```python
plt.plot(stock_data['Daily_Return'],marker='o')
plt.xticks(stock_data.index,rotation=90)
plt.title(symb[:-3]+" Daily Return (%) From "+start_date +" to "+end_date)
plt.grid()
plt.show()
```
